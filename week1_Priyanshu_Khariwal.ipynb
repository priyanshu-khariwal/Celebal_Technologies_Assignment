### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
A = np.array([[4, 1],
              [2, 3]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)  # YOUR CODE HERE

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]

    Av = A @ v
    lv = lam * v

    print(f"\nEigenpair {i+1}:")
    print("A @ v =", Av)
    print("λ * v =", lv)

    # Verify numerically
    assert np.allclose(Av, lv), "Eigenvector equation failed!"

# Plot
fig, ax = plt.subplots()

for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]

    # Scale eigenvector by eigenvalue
    scaled_v = lam * v

    ax.quiver(
        0, 0,
        scaled_v[0], scaled_v[1],
        angles='xy',
        scale_units='xy',
        scale=1
    )

ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Eigenvectors Scaled by Eigenvalues')
ax.axhline(0, color='k', lw=0.8)
ax.axvline(0, color='k', lw=0.8)

plt.grid(True)
plt.tight_layout()
plt.show()

print("4.3 passed ✓")
````

**Your Geometric Explanation (4.3):**
Geometrically, a matrix transforms vectors by stretching, shrinking, rotating, or reflecting them. An eigenvector is a special vector whose direction does not change under the transformation — it only gets scaled. The corresponding eigenvalue tells how much stretching or shrinking occurs. If the eigenvalue is greater than 1, the vector is stretched; if it is between 0 and 1, it shrinks; if it is negative, the vector flips direction in addition to scaling.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X, full_matrices=False)  # YOUR CODE HERE

# Reconstruct X from SVD
X_reconstructed = U @ np.diag(S) @ Vt  # YOUR CODE HERE

# Rank-1 approximation (largest singular value/vector only)
X_approx = S[0] * np.outer(U[:, 0], Vt[0, :])  # YOUR CODE HERE

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

The SVD decomposition is:

X=UΣV^T

SVD decomposes a matrix into orthogonal directions of variation (U and Vᵀ) and their importance (singular values in Σ). PCA is closely related because principal components correspond to the directions of maximum variance, which are given by the right singular vectors (columns of V) of centered data, while the singular values determine how much variance each component explains.

### 5.2 Hypothesis Testing

**Scenario:** You suspect that Engineering salaries are significantly higher than the company average.

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller
from scipy import stats

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)
ks_s1 = stats.kstest(s1, 'norm')  # YOUR CODE HERE
ks_s2 = stats.kstest(s2, 'norm')  # YOUR CODE HERE

print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)  # YOUR CODE HERE — adfuller(ts)

print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts)  # YOUR CODE HERE — np.diff
adf_diff = adfuller(ts_diff)  # YOUR CODE HERE

print(
    f"\nAfter differencing — p-value: "
    f"{adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}"
)
````

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    epsilon = 1e-10

    # Create bins from expected distribution
    breakpoints = np.histogram_bin_edges(expected, bins=bins)

    # Histogram counts
    expected_counts, _ = np.histogram(expected, bins=breakpoints)
    actual_counts, _ = np.histogram(actual, bins=breakpoints)

    # Convert to percentages
    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    # Add epsilon to avoid divide-by-zero/log(0)
    expected_pct = expected_pct + epsilon
    actual_pct = actual_pct + epsilon

    # Compute PSI
    psi = np.sum(
        (actual_pct - expected_pct) *
        np.log(actual_pct / expected_pct)
    )

    return psi


np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)

print(f"PSI: {psi_value:.4f}")
print(
    f"Shift severity: "
    f"{'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}"
)

# Plot both distributions
fig, ax = plt.subplots()

ax.hist(train_dist, bins=30, density=True, alpha=0.6, label='Training')
ax.hist(drift_dist, bins=30, density=True, alpha=0.6, label='Production (Shifted)')

ax.legend()
ax.set_title('Distribution Shift Visualization')

plt.tight_layout()
plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**
The PSI formula is:

∑(actual_pct−expected_pct)ln(
expected_pct/
actual_pct
)

Concept drift: A change in the relationship between inputs and the target variable over time, causing model predictions to become less accurate.

Covariate drift: A change in the distribution of input features while the relationship between inputs and target may remain unchanged.

PSI thresholds: PSI < 0.1 generally indicates no meaningful shift (stable), 0.1–0.2 suggests a minor/moderate shift that should be monitored, and PSI > 0.2 indicates a major shift that may require intervention.

One trigger for retraining: Retrain a model when sustained drift is detected (e.g., PSI > 0.2) or when production performance metrics significantly degrade.

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = (
    p_free_given_spam * p_spam
    + p_free_given_ham * p_ham
)  # YOUR CODE HERE

# Bayes
p_spam_given_free = (
    p_free_given_spam * p_spam
) / p_free  # YOUR CODE HERE

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    p_ham = 1 - prior_spam

    # Evidence P(word)
    p_word = (
        p_word_given_spam * prior_spam
        + p_word_given_ham * p_ham
    )

    # Posterior P(spam | word)
    return (p_word_given_spam * prior_spam) / p_word


pred = naive_bayes_predict(0.30, 0.80, 0.05)

assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````


## Bayes Term Mapping (6.3)

| Term | Formula | In this example |
|------|---------|-----------------|
| Prior | P(Spam) | P(Spam) = 0.30 — probability an email is spam before seeing the word |
| Likelihood | P(word \| Spam) | P("free" \| Spam) = 0.80 — probability of seeing **"free"** given the email is spam |
| Evidence | P(word) | P("free") = P("free" \| Spam) × P(Spam) + P("free" \| Not Spam) × P(Not Spam) = 0.275 |
| Posterior | P(Spam \| word) | P(Spam \| "free") = (P("free" \| Spam) × P(Spam)) / P("free") ≈ 0.8727 |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30

# YOUR CODE HERE — array of 5000 sample means
sample_means = np.array([
    np.mean(np.random.choice(population, size=sample_size))
    for _ in range(n_samples)
])

pop_mean = population.mean()
pop_std  = population.std()

# YOUR CODE HERE — predicted std of sample means
clt_std = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()

# YOUR CODE HERE — histogram of sample means
ax.hist(sample_means, bins=40, density=True, alpha=0.6, label='Sample Means')

# Theoretical normal overlay predicted by CLT
x = np.linspace(sample_means.min(), sample_means.max(), 300)
ax.plot(
    x,
    stats.norm.pdf(x, loc=pop_mean, scale=clt_std),
    label='CLT Normal Approximation'
)

ax.set_title('CLT: Distribution of Sample Means')
ax.legend()

plt.tight_layout()
plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))

print(
    f"KS test p-value: {ks_result.pvalue:.4f} "
    f"→ Approximately normal? {ks_result.pvalue > 0.05}"
)
````

## CLT Reflection (6.4)

The Central Limit Theorem matters because it shows that the distribution of sample means becomes approximately normal, even when the original population distribution is not normal, provided the sample size is sufficiently large. This allows statisticians and machine learning practitioners to make reliable inferences about populations using sample statistics. The CLT justifies the normality assumption used in many classical statistical tests, such as t-tests, confidence intervals, and linear regression inference.
